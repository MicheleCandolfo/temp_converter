```python
#Definition und Benennung der Funktion
#value_in = Zahl der umzurechnenden Temperatureinheit
#unit_in = Ausgangs-Temparatureinheit
#unit_out = Gewünschte Temperatureinheit
def conv_temperature(value_in, unit_in, unit_out):
    """ Rechnet die Temperatur zwischen Kelvin (K), Grad Delisle (°D), Grad Celsius (°C), Grad Réaumur (°Re), 
    Grad Fahrenheit (°F) und Grad Rankine (°Ra) um. (value_in) stellt 
    die Temperatureingabe in der spezifizierten Einheit (unit_in) dar und soll in die angebene Einheit (unit_out) 
    konvertiert werden."""

    #Ausnahmebehandlung: Check, ob value_in vom Typ int oder float ist. 
    #Bei fehlerhafter Prüfung erscheint eine Fehlermeldung.
    try:
        valuecheck_int = isinstance(value_in,int)
        valuecheck_float = isinstance(value_in,float)
        assert valuecheck_int or valuecheck_float == True  
    
    except AssertionError:
        print("Datentyp bei Value_in falsch. Bitte integer oder float eingeben")
    
    #Ausnahmebehandlung: Check, ob unit_in vom Typ str ist.
    #Bei fehlerhafter Prüfung erscheint eine Fehlermeldung.
    else:
        try:
            unit_in_check = isinstance(unit_in,str)
            assert unit_in_check == True
        except AssertionError:
            print("Datentyp bei Unit_in falsch. Bitte String eingeben. \nMögliche Werte: k (Kelvin),c (Celsius),f (Fahrenheit), ra (Rankine), re (Reaumur),d (Delisle).")
        else:
    
    #Ausnahmebehandlung: Check, ob unit_out vom Typ str ist.
    #Bei fehlerhafter Prüfung erscheint eine Fehlermeldung.        
            try:
                unit_out_check = isinstance(unit_out,str)
                assert unit_out_check == True
            except AssertionError:
                print("Datentyp bei Unit_out falsch. Bitte String eingeben. \nMögliche Werte: k (Kelvin),c (Celsius),f (Fahrenheit), ra (Rankine), re (Reaumur),d (Delisle).")
            
            #Sobald die Prüfungen bestanden wurden, wird der weitere Code ausgeführt.
            else:
                
                #Globalsetzen der objekte, damit der Typ außerhalb der Funktion im global Environment geprüft werden kann. Dadruch wird das Debugging erleichtert.  
                global value_in_controll  
                value_in_controll = value_in
                global unit_in_controll
                unit_in_controll = unit_in
                global unit_out_controll
                unit_out_controll = unit_out
                
                #Sichert die Eingabe kleiner Buchstaben, auch wenn große eingetragen werden.
                unit_in = unit_in.lower()
                unit_out = unit_out.lower()
                
                #Listen, um korrekte Eingabe der unit_in und -out zu überprüfen. 
                #Bei nichtbestehen erscheint Fehler/-Hilfemeldung.
                list_unit_check =["c","f","ra","k","d","re"]
            
                #Überprüfung, ob eingegebener Wert in der Prüfliste existiert.
                if unit_in not in list_unit_check or unit_out not in list_unit_check:
                    print("Die angegebenen Werte werden nicht untersützt. Bitte möglichen Wert eingeben und Ausführung wiederholen.\nMögliche Werte: k (Kelvin),c (Celsius),f (Fahrenheit), ra (Rankine), re (Reaumur), d (Delisle)")

                #Überprüfung ob unit_in und unit_out identisch sind.   
                elif unit_in == unit_out:
                    print("Die angegebenen Werte für unit_in und unit_out sind identisch, bitte nochmals überprüfen.")

                #Überprüfung der Einhaltung der physikalischen Gesetze, Eingaben unter dem absoluten Nullpunkt der spezifischen Temperatureinheit werden nicht konvertiert.
                elif (unit_in == 'k' or unit_in == 'ra') and value_in < 0:
                    print('Die Eingabe befindet sich unter den absoluten Nullpunkt und ist nicht zulässig, bitte geben Sie einen Wert größer gleich 0 ein.')

                elif unit_in == 'c' and value_in <  -273.15:
                    print('Die Eingabe befindet sich unter den absoluten Nullpunkt und ist nicht zulässig, bitte geben Sie einen Wert größer gleich -273.15 ein.')

                elif unit_in == 'f' and value_in <  -459.67:
                    print('Die Eingabe befindet sich unter den absoluten Nullpunkt und ist nicht zulässig, bitte geben Sie einen Wert größer gleich -459.67 ein.')
                
                elif unit_in == 're' and value_in <  -218.52:
                    print('Die Eingabe befindet sich unter den absoluten Nullpunkt und ist nicht zulässig, bitte geben Sie einen Wert größer gleich -218.52 ein.')
                
                elif unit_in == 'd' and value_in <  559.72:
                    print('Die Eingabe befindet sich unter den absoluten Nullpunkt und ist nicht zulässig, bitte geben Sie einen Wert größer gleich 559.72 ein.')
                           
                #Wenn die Eingaben in korrektem Typ und Form sind, beginnt die Konvertierung.
                else:

                    # Celsius wird als Basiseinheit für weitere Berechnungen/Konvertierungen definiert.
                    if unit_in  == 'c' or unit_in == 'k' or unit_in == 'f' or unit_in == 'ra' or unit_in == 're' or unit_in == 'd':
                        # Umrechnung von Kelvin auf Celsius
                        if unit_in == 'k':
                            value_in = value_in - 273.15
                        # Umrechnung von Fahrenheit auf Celsius
                        if unit_in == 'f':
                            value_in = (value_in - 32) * 5/9
                        # Umrechnung von Rankine auf Celsius
                        if unit_in == 'ra':
                            value_in = (value_in * 5/9) - 273.15
                        # Umrechnung von Reaumur auf Celsius
                        if unit_in == 're':
                            value_in = value_in * 1.25
                        # Umrechnung von Delisle auf Celsius
                        if unit_in == 'd':
                            value_in = 100 - (value_in * 2/3)
                    #-----------------------------------------------
                    # Berechnung/Konvertierung in Zieleinheit auf der Basis Celsius.
                        if unit_out == 'f':
                            value_out = (value_in * 1.8) + 32

                        elif unit_out == 'k':
                            value_out = value_in + 273.15

                        elif unit_out == 'c':
                            value_out = value_in              

                        elif unit_out == 'ra':
                            value_out = (value_in * 1.8) + 491.67

                        elif  unit_out == 're':
                            value_out = value_in * 0.8

                        elif unit_out == 'd':
                            value_out = (100 - value_in) * 1.5           
                    
                    #Zuletzt wird das Objekt value_out global gesetzt, da es erst im Laufe der Funktion generiert wird.
                    global value_out_controll
                    value_out_controll = value_out  
                    
                    #Gibt den Funktionswert in float aus.
                    return float(value_out)
```
